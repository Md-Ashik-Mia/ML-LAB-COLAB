```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/diabetes.csv")
df.head()
```

```python
x=df.iloc[:,:-1]
print(x)
```

```python
for column in x :
  # print(df[column].mean())
  # print(df[column])
    df[column]=df[column].replace(0,df[column].mean())
df.head()
# Replace zero values in key features with column mean or median. //done here
```

```python
highestGlucoseValue =df["Glucose"].max()
df.iloc[0,1]=highestGlucoseValue
df.head()
# Replace the first row’s glucose value with the maximum from the glucose column. //done here
```

```python
minimumAge =df["Age"].min()
minimumGlucose =df["Glucose"].min()
print(minimumAge,minimumGlucose)
```

```python
# for column in x :
  # if column=="Age":
index=0
for i in df["Age"]:
    print(index, i)
    if(i==minimumAge):
      df.iloc[index,1]=minimumGlucose
    index=index+1
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, train_size=0.2, random_state=0)
# preprocessing data here
```

```python
from re import RegexFlag
from sklearn.linear_model import LinearRegression
regressor =LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
import numpy as np
y_pred=regressor.predict(x_test)
y_pred=np.round(y_pred)
# Use Linear Regression to predict the binary Outcome. Round predictions to the nearest integer (0 or 1). //done here

updatedf=pd.DataFrame({"Actual":y_test , "Predict":y_pred})
print(updatedf)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix,precision_score,recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
confusion=confusion_matrix(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
fscore = f1_score(y_test,y_pred)
print(accuracy)
print(confusion)
print(precision)
print(recall)
print(fscore)
```
